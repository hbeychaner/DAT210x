```python
def getBestScoreIso(u, o):
  for p in range(2, 6):
    for r in range(4, 7):
      model_score = 0
      best_score = 0
      best_combo = [0] * 6
      iso = ISO(n_neighbors=p, n_components = r)
      u_norm = iso.fit_transform(u)
      o_norm = iso.fit_transform(o)
    for i in c:
      for j in g:
        svc = SVC(gamma=j, C=i)
        svc.fit(u_norm, y_train)
        model_score = svc.score(o_norm, y_test)
        if model_score > best_score:
          best_score = model_score
          best_combo[0] = model_score
          best_combo[1] = (f"C: {i}")
          best_combo[2] = (f"Gamma: {j}")
          best_combo[3] = (f"PreProc: {h}")
          best_combo[4] = (f"n_neighbors: {p}")
          best_combo[5] = (f"n_components: {r}")
          print(best_combo)
  return(best_combo)
```
